```python
import pandas as pd
```

# read data

```python
roberta_df = pd.read_csv("tom_scott_copyright_sentiment_roberta.csv")
roberta_df.head(5)
```

```python
vader_df = pd.read_csv('tom_scott_copyright_sentiment_vader.csv')
vader_df.head(5)
```

## combine data together

```python
vader_scores = vader_df[['vader_neg','vader_neu','vader_pos','vader_compound']]
vader_scores.head()
```

```python
compiled_data = roberta_df[["snippet.textOrginal"]]
compiled_data = pd.concat([compiled_data, roberta_df[["roberta_neg","roberta_neu","roberta_pos"]], vader_scores], axis=1)
compiled_data.head()
```

# sort data

```python
roberta_sorted = compiled_data.query('roberta_neg > 0.5').sort_values('roberta_neg', ascending=False).reset_index(drop=True)
roberta_sorted.to_csv("sorted_by_roberta.csv", index=False)
roberta_sorted.head(5)
```

```python
vader_sorted = compiled_data.query('vader_compound < -0.5').sort_values('vader_compound').reset_index(drop=True)
vader_sorted.to_csv("sorted_by_vader.csv", index=False)
vader_sorted.head(5)
```
